````python
locations = {
    "admu":{
        "name":"Ateneo de Manila University",
        "latitude":14.6394557,
        "longitude":121.0758975        
    },
    "upd":{
        "name":"University of the Philippines-Diliman Campus",
        "latitude":14.6538519,
        "longitude":121.0663348
    },
    "dlsu":{
        "name":"De La Salle University-Taft",
        "latitude":14.5647346,
        "longitude":120.9909876
    }
}

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

def eta(source,destination):
    time = 0
    if source != destination:
        if (source,destination) in legs:
            time = legs[source,destination]['travel_time_mins']
        else:            
            newList = list(legs.keys())
            sourceList = []
            destinationList = []
            for i in newList:
                sourceList.append(i[0])
                destinationList.append(i[1])
                
            time = legs[newList[sourceList.index(source)]]['travel_time_mins'] + legs[newList[destinationList.index(destination)]]['travel_time_mins']
    else:
        return 0;
            
    return time

print(eta('admu','upd'))
print(eta('dlsu','admu'))

source = input("Enter source: ")
destination = input("Enter destination: ")
eta(source,destination)
````
